```python
import numpy as np
import matplotlib.pyplot as plt
import os
from PIL import Image
```

```python
path = os.path.join("C:\\Users\\Robin\\Desktop\\deep_learning")
#video number 1
file = np.load(path+"\\processed_data\\v3_absolute_grape_changeling-12_2293-4124NDTRAIN.npy", allow_pickle=True)[1]
#first five frames of the video
file = np.array(file[-1792:][:5]).astype(np.uint8)

#the first frame's RGB value subtracting the 2nd video
difference = file[4]-file[0]
difference = Image.fromarray(difference, "RGB")
difference = difference.convert("LA")
#results
print(difference)

#image of the results
plt.imshow(difference)
plt.show()
plt.imshow(file[0])
plt.show()
plt.imshow(file[4])
plt.show()
```
